# Setup

```python
import geopandas as gpd
import osmnx as ox
import folium
import pandas as pd
from shapely.geometry import Point
```

# World boundaries

```python
world = gpd.read_file('./boundaries/boundaries.shp')
```

```python
world.head()
```

# Pitches

## Polygon

```python
country_of_choice = 'United Kingdom'
cc = 'uk'
```

```python
country = world[world.ADMIN == country_of_choice]
```

```python
poly = country.geometry.iloc[0]
```

## Query OSM

```python
# Download all leisure footprints in our focus area
leisure = ox.footprints.footprints_from_polygon(poly, footprint_type='leisure')
```

## Filter for soccer pitches

```python
pitches = leisure[(leisure.leisure=='pitch') | (leisure.leisure=='stadium')]
```

```python
soccer = pitches[pitches.sport == 'soccer']
```

```python
pd.options.display.max_columns = 300
```

```python
soccer.info()
```

```python
soccer.head()
```

## Plot

```python
soccer.loc[:,['geometry', 'leisure', 'note', 'name', 'surface']].to_file("./geojson/osm_soccer_pitches_" + cc + ".geojson", driver='GeoJSON')
```

```python
soccer_x = [point.x for point in soccer.geometry.centroid]
soccer_y = [point.y for point in soccer.geometry.centroid]
```

```python
tileset_ID_str = "satellite-v9"
mapbox_access_token = "pk.eyJ1Ijoic2VieHdvbGYiLCJhIjoiY2pvbmVlbHVvMHV5cjNrcnl3ZnU0cGU3eSJ9.jP3kUXiKwKFXwQyUgutldQ"
tilesize_pixels = '512'
mapbox_tile_URL = f"https://api.mapbox.com/styles/v1/mapbox/{tileset_ID_str}/tiles/{tilesize_pixels}/{{z}}/{{x}}/{{y}}@2x?access_token={mapbox_access_token}"
```

```python
m = folium.Map(
        location=[53, 0],
        tiles=None,
        zoom_start=8,
        control_scale = True)

folium.TileLayer('openstreetmap', name = 'OpenStreetMap').add_to(m)
folium.TileLayer(mapbox_tile_URL, attr = 'Mapbox', name = 'Satellite', overlay = False).add_to(m)

folium.GeoJson(soccer, name = 'Football Pitches', 
        style_function= lambda x: {"weight": 2, 'color': 'black', 'fillColor': 'blue', 'fillOpacity':0.8},
        highlight_function=lambda x: {'weight':4, 'color':'black'},
        smooth_factor=1,
        show = True,
        tooltip=folium.features.GeoJsonTooltip(fields = ['name', 'note', 'surface' , 'source',],
                                                  aliases=['Name', 'Note' , 'Surface', 'Source'], 
                                                  labels=True, 
                                                  sticky=True,
                                                  localize=True
                                                 )).add_to(m)

soccer_list = zip(soccer_y, soccer_x)
for lat, lon in soccer_list:
    folium.CircleMarker(location = [lat, lon], 
                        fill_color = 'green',
                        radius = 5,
                        color='g',
                        fill = True,
                        fill_opacity = 0.7).add_to(m)

folium.LayerControl().add_to(m)
    
m.save('./html_maps/osm_soccer_pitches_' + cc + '.html')
```

```python
world_low_res
```

```python
europe = world_low_res[world_low_res.continent == 'Europe']
```

```python
europe_poly = europe.geometry.iloc[0]
```

### UEFA stadium list

```python
stadiums = pd.read_excel('./uefa/stadiums_sample_UEFA-Datatonic.xlsx')
```

```python
stadium_location_list = zip(stadiums.latitude, stadiums.logitude, )
```

```python
stadiums['geometry'] = [Point(point[1], point[0]) for point in stadium_location_list]
```

```python
stadiums_gpd = gpd.GeoDataFrame(stadiums, geometry='geometry', crs = europe.crs)
```

```python
stadiums_gpd_uk = gpd.sjoin(stadiums_gpd, uk, how = "inner", op = 'within') 
```

```python
stadium_location_list = zip(stadiums.latitude, stadiums.logitude)
soccer_list = zip(soccer_y, soccer_x)

m = folium.Map(
        location=[53, 0],
        tiles=None,
        zoom_start=5,
        control_scale = True)

folium.TileLayer('openstreetmap', name = 'Map').add_to(m)
folium.TileLayer(mapbox_tile_URL, attr = 'Mapbox', name = 'Satellite', overlay = False).add_to(m)
    
for lat, lon in stadium_location_list:
    folium.CircleMarker(location = [lat, lon], 
                        fill_color = 'red',
                        radius = 10,
                        color='r',
                        fill = True,
                        fill_opacity = 1).add_to(m)
    
for lat, lon in soccer_list:
    folium.CircleMarker(location = [lat, lon], 
                        fill_color = 'green',
                        radius = 5,
                        color='g',
                        fill = True,
                        fill_opacity = 0.7).add_to(m)

folium.LayerControl().add_to(m)
    
m.save('./html_maps/osm_uefa_comparison_uk.html')
```

### De-duplicate

```python
stadiums_uk_in_osm = gpd.sjoin(stadiums_gpd_uk_dedupe, soccer, how = "inner", op = 'within') 
```
